Обучить модель с экспоненциальным ростом числа заболевших:

$$y \sim \exp(linreg) \Leftrightarrow \ln{y} \sim linreg$$

```python
import numpy as np

class MyLinReg:
    def __init__(self, X, Y, exp=True, free_term=True):
        self.free_term = free_term
        if exp:
            self.Y = np.log(np.array(Y).T)
        else:
            self.Y = np.array(Y).T
        if self.free_term:
            self.X = np.c_[np.array(X), np.ones(len(X))]
        else:
            self.X = np.array(X).reshape(-1, 1)
        self.a = None
    def fit(self):
        self.a = np.linalg.inv(self.X.T.dot(self.X)).dot(self.X.T).dot(self.Y)
    def predict(self, X):
        if self.free_term:
            X = np.c_[np.array(X), np.ones(len(X))]
        return np.sum(X * self.a, axis=1)
```

```python
lr = MyLinReg(data_train['day_number'].values,
              data_train['total_cases'].values)
lr.fit()
lr1_pred = lr.predict(data_test['day_number'].values)
```

```python
# сравнение с методом из коробки
from sklearn.linear_model import LinearRegression

lr_2 = LinearRegression()
fitted = LinearRegression().fit(np.array(data_train['day_number']).reshape(-1, 1),
              np.log(data_train['total_cases'].values))
lr2_pred = fitted.predict(data_test['day_number'].values.reshape(-1, 1))
```

```python
# результаты предикта одинаковые

eps = 10 ** (-10)
assert sum(lr1_pred - lr2_pred) < eps
```
